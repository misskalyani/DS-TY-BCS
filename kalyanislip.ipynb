#### Q.2 A) Write a Python program to create a graph to find relationship between the petal length and petal width.(Use iris.csv dataset)

```python
import pandas as pd
df=pd.read_csv("/home/kalyani/DS/Species/Iris.csv")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#plt.figure(figsize=(8, 6))
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', data=df, hue='Species')
plt.title('Relationship between Petal Length and Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='PetalLengthCm', y='PetalWidthCm', data=df, scatter=True)
plt.title('Relationship between Petal Length and Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

#### Q.2) Dataset Name: winequality-red.csv
#### Write a program in python to perform following tasks
#### a. Rescaling: Normalised the dataset using MinMaxScaler class
#### b. Standardizing Data (transform them into a standard Gaussian distribution with a mean of 0 and  a standard deviation of 1)
#### c. Normalizing Data ( rescale each observation to a length of 1 (a unit norm). For this, use the Normalizer class.)

```python
import pandas as pd
df=pd.read_csv("/home/kalyani/DS/wineequality-red/winequality-red.csv")
df
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Normalizer

# Load Wine Quality dataset
df = pd.read_csv('/home/kalyani/DS/wineequality-red/winequality-red.csv')

# Split dataset into features (X) and target (y)
X = df.drop('quality', axis=1)
y = df['quality']

# Rescaling using MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
print("Rescaled Data (MinMaxScaler):")
print(pd.DataFrame(X_scaled, columns=X.columns).head())

# Standardizing Data using StandardScaler
std_scaler = StandardScaler()
X_std = std_scaler.fit_transform(X)
print("\nStandardized Data (StandardScaler):")
print(pd.DataFrame(X_std, columns=X.columns).head())

# Normalizing Data using Normalizer
norm_scaler = Normalizer()
X_norm = norm_scaler.fit_transform(X)
print("\nNormalized Data (Normalizer):")
print(pd.DataFrame(X_norm, columns=X.columns).head())
```
