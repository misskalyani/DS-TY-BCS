```python
#Create Dataframe using numpy array
import numpy as np
data=np.array([['Tata','Nexon',2017],['KIA','Seltos',2019],['Hyundai','Creta',2015]])
df=pd.DataFrame(data,columns = ['company','model','year'])
df
```

```python
#Each element of the list is one record in the dataframe.the column headers are passed seprately
import pandas as pd
data=[['Tata','Nexon',2017],['MG','Astor',2021],['KIA','Seltos',2019],['Hyundai','Creta',3015]]
df=pd.DataFrame(data,columns = ['comapny','model','year'])
df
```

```python
#Create dataframe using dictionary with lists
import pandas as pd
data ={'comapny':['Tata','MG','KYA','Hyundai'],
       'model':['Nexon','Astor','Creta','Seltos'],
        'year':[2017,2021,2019,2015]}
df=pd.DataFrame.from_dict(data)
df
```

```python
# create datafram using list of dictionary
import pandas as pd
data=[{'company':'Tata','model':'nexon','year':2017},
     {'company':'MG','model':'Astor','year':2021},
     {'company':'KIA','model':'seltos','year':2019},
     {'company':'Hyundai','model':'Creta','year':2015}]
df = pd.DataFrame(data)
df
```

```python
#create dataframe using zip() function
import pandas as pd
company=['Tata','MG','KIA','Hyundai']
model=['Nexon','Astor','Seltos','Creta']
year=[2017,2021,2019,2015]
data=list(zip(company,model,year))
df=pd.DataFrame(data,columns =['company','model','year'])
df
```

```python
#Adding rows and columns with invalid,duplicate or missing data
import pandas as pd
df=pd.DataFrame(columns = ['comapny','model','year'])
df.loc[0]=['Tata','Nexon',2017]
df.loc[1]=['MG','Astor',2021]
df.loc[2]=['KIA','Seltos',2019]
df.loc[3]=['Hyundai','Creta',2015]
df.loc[4]=['Honda','Jazz',None]
df.loc[5]=[None,None,None]
df.loc[6]=['Toyota',None,2018]
df.loc[7]=['Tata','Nexon',2017]
df["newcolumn"]=None
df
```
